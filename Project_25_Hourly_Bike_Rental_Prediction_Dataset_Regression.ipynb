```python
#Lets seprate the input and ouput variable
df_x=df_bikes.drop(columns=["casual","registered","cnt","dteday"])
y=df_bikes[["cnt"]]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(df_x)
x=sc.transform(df_x)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
def maxr2_score(regr,df_x,y):
    max_r_score=0
    for r_state in range(42,100):
        x_train, x_test, y_train, y_test = train_test_split(df_x, y,random_state = r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        y_pred = regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print("r2 score corresponding to ",r_state," is ",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
    return final_r_state
```

```python
#lets make a function which evaluates the model using cross_val_score
from sklearn.model_selection import cross_val_score
def model_evaluation(model,x,y):
    c_scores=cross_val_score(model,x,y,cv=5,scoring="r2")
    print("Mean r2 score for regressor: ",c_scores.mean())
    print("standard deviation in r2 score for regressor: ",c_scores.std())
    print(c_scores)
```

```python
#using Decision Trees
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
maxr2_score(dtr,df_x,y)
```

```python
#Lets chcek KNN
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
knr=KNeighborsRegressor()
parameters = {"n_neighbors":range(2,30)}
clf = GridSearchCV(knr, parameters, cv=5,scoring="r2")
clf.fit(x, y)
clf.best_params_
```

```python
knr=KNeighborsRegressor(n_neighbors=16)
maxr2_score(knr,x,y)
```

```python
#lets use ensemble model Random forets
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings("ignore")
rfr=RandomForestRegressor()
parameters = {"n_estimators":[10,100,500]}
clf = GridSearchCV(rfr, parameters, cv=5,scoring="r2")
clf.fit(df_x,y)
clf.best_params_
```

```python
rfr=RandomForestRegressor(n_estimators=500)
maxr2_score(rfr,df_x,y)
```

```python
#Lets chcek the cross val scores of above models
```

```python
print("Decision Tree\n\n")
model_evaluation(dtr,df_x,y)
```

```python
print("Random Forest\n\n")
model_evaluation(rfr,df_x,y)
```

```python
#Based on above scores we will choose random forest as our final algo
x_train, x_test, y_train, y_test = train_test_split(df_x, y,random_state = 91,test_size=0.20)
rfr.fit(x_train,y_train)
y_pred = rfr.predict(x_test)
```

```python
#Lets find the rmse and r2_score using sklearn.metrics for random forest model
print("MSE is: ",mean_squared_error(y_test,y_pred))
print("r2_score is: ",r2_score(y_test,y_pred))
```
