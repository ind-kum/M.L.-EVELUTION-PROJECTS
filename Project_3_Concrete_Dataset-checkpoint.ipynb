```python
#Treating using Log
import numpy as np
for col in df_concrete.columns:
    if df_concrete.skew().loc[col]>0.55:
        df_concrete[col]=np.log1p(df_concrete[col])
        
df_concrete.skew()
```

```python
#Lets check for outliers
from scipy.stats import zscore
z_score=abs(zscore(df_concrete))
print(df_concrete.shape)
df_concrete_final=df_concrete.loc[(z_score<3).all(axis=1)]
print(df_concrete_final.shape)
```

```python
#seprating target and imput variables
#np.logip
import numpy as np
df_x=df_concrete_final.drop(columns=["Concrete compressive strength"])
y=df_concrete_final[["Concrete compressive strength"]]
```

```python
#scaling the input variable
#Linearregression algorithm requires all features to be on common scale
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = 55,test_size=0.20)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
max_r_score=0
for r_state in range(42,100):
    x_train, x_test, y_train, y_test, = train_test_split(x, y,random_state = r_state,test_size=0.20)
    regr = linear_model.LinearRegression()
    regr.fit(x_train,y_train)
    y_pred = regr.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(linear_model.LinearRegression(),x,y,cv=10,scoring="r2").mean()
```
